```python
# importing pandas module
import pandas as pd
```

```python
#Load Sheets
xl = pd.ExcelFile('nutrients.xlsx')
```

```python
dfmain = pd.DataFrame(columns=["Food","Portion","Unit","iron-mg", "Potassium-mg", "magnesium-mg", "phosphorus-mg",
                               "zinc-mg", "manganese-mg", "fluoride-mcg", "chromium-mcg", "chlorine-mcg", "selenium-mcg", 
                               "molybdenum-mcg", "iodine-mcg", "calcium-mg", "copper-mg", "sodium-mg", "vitamin-C-mg", 
                               "vitamin-E-mg", "Thiamin-B1-mg", "riboflavin-B2-mg", "niacin-B3-mg", "pantothenic-Acid-B5-mg", 
                               "vitamin-B6-mg", "biotin-mcg", "folate-B9-mcg", "vitamin-B12-mcg", "Vitamin-D-mcg", 
                               "vitamin-K-phylloquinone-mcg"])
```

```python
more = "yes"

print("Categories:")
for i in range(0, len(xl.sheet_names)):
    print(xl.sheet_names[i])

while more != 'no':
    
    category = input("Please choose a category: ")

    # Convert the sheet name to a title.
    category = category.title()

    #Load sheet
    data = pd.read_excel("nutrients.xlsx", category)

    # Converts the names of food to a list.
    names = data["Food"].tolist()
    
    name = input("Enter a name of food: ")

    # Converts the name to a title to match the names of foods.
    name = name.title()

    # Prints out the foods corresponding to the given name of food.
    for i in range(0, len(names)):
        if name in names[i]:
            print(str(i) + ") " + names[i])
    
    response = input("Does your food match one of these options: ")

    if response == "yes" or response == "Yes" or response == "YES":
        number = input("Enter the number corresponding to your food: ")
    else:
        print("Please narrow down your search")
        print("enter additional description")
        additionalNames = input()
        txt = additionalNames.split()
        
    df1 = data.iloc[int(number):int(number)+1,:]
    df1 = df1.fillna(0)
    df1.head()
    
    df1["Portion"] = pd.to_numeric(df1["Portion"])
    
    portions = float(input("enter the number of portions: "))
    
    df1.iloc[:,3:] = df1.iloc[:,3:].mul(portions)
    df1.iloc[:,1:2] = df1.iloc[:,1:2].mul(portions)

    df1.head()
    
    dfmain = dfmain.append(df1)

    dfmain = dfmain.fillna(0)

    dfmain.head()
    
    more = input("Do you have more food to add? (yes or no): ")
    
dfmain.to_excel("output.xlsx", index=False)
```
